```python
import random
import statistics

kitaplar = [
 {"isim": "Veri Bilimi 101", "yazar": "Ali", "tur": "Bilim", "satis": 1200, "yil": 2021},
 {"isim": "Python ile Yapay Zeka", "yazar": "Ayşe", "tur": "Bilim", "satis": 950, "yil": 2020},
 {"isim": "İstatistik Temelleri", "yazar": "Ali", "tur": "Akademik", "satis": 700, "yil": 2019},
 {"isim": "Makine Öğrenmesi", "yazar": "Can", "tur": "Bilim", "satis": 1800, "yil": 2022},
 {"isim": "Veri Görselleştirme", "yazar": "Deniz", "tur": "Sanat", "satis": 400, "yil": 2018},
 {"isim": "Matematiksel Modelleme", "yazar": "Ali", "tur": "Akademik", "satis": 1500, "yil": 2021},
 {"isim": "Bilgi Toplumu", "yazar": "Ayşe", "tur": "Sosyal", "satis": 600, "yil": 2022}
]

def en_cok_satan(kitaplar):
    return max(kitaplar, key=lambda k: k['satis'])

def yazar_satislari(kitaplar):
    toplam = {}
    for k in kitaplar:
        toplam[k['yazar']] = toplam.get(k['yazar'], 0) + k['satis']
    return toplam

turler = set([k['tur'] for k in kitaplar])
yuksek_satis_kitaplar = [k['isim'] for k in kitaplar if k['satis'] > 1000]

sonra_kitaplar = list(filter(lambda k: k['yil'] > 2020, kitaplar))
yeni_satislar = list(map(lambda k: k['satis']*1.1, kitaplar))
siralama = sorted(kitaplar, key=lambda k: k['satis'], reverse=True)

ortalama_satis = statistics.mean([k['satis'] for k in kitaplar])
tur_satis = {}
for k in kitaplar:
    tur_satis[k['tur']] = tur_satis.get(k['tur'], 0) + k['satis']
en_cok_satan_tur = max(tur_satis, key=tur_satis.get)
satis_std = statistics.stdev([k['satis'] for k in kitaplar])

train_size = int(len(kitaplar)*0.7)
train = random.sample(kitaplar, train_size)
test = [k for k in kitaplar if k not in train]

train_yazar_ortalama = {}
for k in train:
    train_yazar_ortalama[k['yazar']] = train_yazar_ortalama.get(k['yazar'], []) + [k['satis']]
for yazar in train_yazar_ortalama:
    train_yazar_ortalama[yazar] = sum(train_yazar_ortalama[yazar])/len(train_yazar_ortalama[yazar])

test_ust_ortalama = [k['isim'] for k in test if k['satis'] > train_yazar_ortalama.get(k['yazar'], 0)]

print("En çok satan kitap:", en_cok_satan(kitaplar)['isim'])
print("Yazar satışları:", yazar_satislari(kitaplar))
print("Tüm türler:", turler)
print("1000'den fazla satan kitaplar:", yuksek_satis_kitaplar)
print("2020 sonrası çıkan kitaplar:", [k['isim'] for k in sonra_kitaplar])
print("Satışlar %10 artırılmış:", yeni_satislar)
print("Satışa göre sıralama:", [k['isim'] for k in siralama])
print("Ortalama satış:", ortalama_satis)
print("En çok satış yapan tür:", en_cok_satan_tur)
print("Satışların standart sapması:", satis_std)
print("Train seti:", [k['isim'] for k in train])
print("Test seti:", [k['isim'] for k in test])
print("Train yazar ortalamaları:", train_yazar_ortalama)
print("Testte ortalamanın üstünde satan kitaplar:", test_ust_ortalama)
```
